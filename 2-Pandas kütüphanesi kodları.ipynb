#Pandas veri yapıları -Seriler
pd.Series(veri,etiket)

```python
urunler=["simit","tost","doner","cay","ayran","su"]
fiyatlar=[10,30,50,5,8,3]
tablo=pd.Series(fiyatlar,urunler)
tablo
```

```python
print("simit ve çay toplam",tablo["simit"]+tablo["cay"],"ytl")
```

```python
konular=["Turkce","Sosyal","matematik","Fen"]
netler=[20,20,20,20]
deneme1=pd.Series(netler,konular)
deneme2=pd.Series([10,10,10,10],konular)
print("Deneme 1:\n ",deneme1)
print(30*"-")
print("Deneme 2 :\n",deneme2)
```

# Pandas veri yapıları -Veri çerçeveleri
pandas serilierini birleştirerek veri çerçevesi oluşur
pd.DataFrame (veriler, satır etiketleri,sütun etiketleri)



```python
konular=["Turkce","Sosyal","Matematik","Fen"]
netler=[[20,20,20,20],[30,20,10,20],[10,20,30,25],[15,10,25,25]]
deneme_isimleri=["deneme1","deneme2","genel tekrar","Türkiye geneli"]
denemeler=pd.DataFrame(netler,deneme_isimleri,konular)
denemeler
```

```python
print("Deneme sınavlarındaki matematik netleri :\n",denemeler["Matematik"])
```

birden çok sütun seçilirken közeli paranta içinde yazmaya dikkat ediniz

```python
print("Deneme sınavlarındaki sosyal ve matematik netleri :\n")
print(denemeler[["Sosyal","Matematik"]])
```

yeni bir sütun oluşturma

```python
denemeler["Toplam"]=denemeler["Turkce"]+denemeler["Sosyal"]+denemeler["Matematik"]+denemeler["Fen"]
denemeler
```

Satır seçme

```python
print("deneme 1 sonuçları :\n",denemeler.loc["deneme1"])
```

```python
print("deneme 1 sonuçları :\n",denemeler.iloc[0])
```

```python
print("deneme 1 ve 2 sonuçları :\n",denemeler.loc[["deneme1","deneme2"]])
```

```python
print("deneme 1 ve 2 sonuçları :\n",denemeler.iloc[[0,1]])
```

```python
print("deneme 1 sosyal sonucu :", denemeler.loc["deneme1","Sosyal"] )
print("deneme 2 sosyal sonucu :", denemeler.loc["deneme2","Sosyal"] )
```

```python
print("deneme 1 ve 2 sosyal sonucu :\n", denemeler.loc[["deneme1","deneme2"],"Sosyal"] )
```

```python
print("deneme 1,2 Matematik ve fen sonucu :\n", denemeler.loc[["deneme1","deneme2"],["Matematik","Fen"]] )
```

```python
print("Türkçe ortalaması    :",denemeler["Turkce"].mean())
print("Sosyal ortalaması    :",denemeler["Sosyal"].mean())
print("Matematik ortalaması :",denemeler["Matematik"].mean())
print("Fen ortalaması       :",denemeler["Fen"].mean() )
```

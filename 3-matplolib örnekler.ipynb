Matplolib kütüphanesini içe aktarma 

```python
import matplotlib.pyplot as plt
import numpy as np
```

grafik çizme

```python
yatay_veriler=np.linspace(0,10,15)
dikey_veriler=yatay_veriler**2
plt.plot(yatay_veriler,dikey_veriler)
plt.show()
```

```python
plt.plot(yatay_veriler,dikey_veriler,"r")
plt.xlabel("x ekseni verileri")
plt.ylabel("y ekseni verileri")
plt.title("grafik başlığı")
plt.show()
```

alt grafikler kullanar akaynı tualde birden çok grafik oluşturma

```python
# satır sayısı sütün ayısı,çizim numarası
plt.subplot(1,2,1) # tek satır 2 çizim birinci çizim
plt.plot (yatay_veriler,dikey_veriler,"r")
plt.subplot(1,2,2) # tek satır 2 çizim ikinci çizim
plt.plot (dikey_veriler,yatay_veriler,"g")
```

```python
# satır sayısı sütün ayısı,çizim numarası
plt.subplot(2,2,1) # iki satır 2 sütun toplam 4 çizim birinci çizim
plt.plot (yatay_veriler,dikey_veriler,"r")
plt.subplot(2,2,4) # iki satır 2 sütun toplam 4 çizim dördüncü çizim
plt.plot (dikey_veriler,yatay_veriler,"g")
```

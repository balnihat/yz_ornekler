#### pandas veri yapıları - Seriler -

```python
pandas_veri= pd.Series([1,2,3,4])
print("pandas_veri \n",pandas_veri)
print("-"*30)
print("ilk veri",pandas_veri[0])
```

```python
pandas_etiketli_veri=pd.Series([1,2,3,4],["A","B","C","D"])
print("etiketli_veri\n",pandas_etiketli_veri)
print("-"*30)
print("b satırı",pandas_etiketli_veri["B"])
```

```python
import numpy as np 
sayılar=np.arange(10,20,2)
print(sayılar[0])
pandas_sayılar=pd.Series(sayılar)
print(pandas_sayılar[0])
etiketler=["A","B","C","D","E"]
pandas_etiketli_sayılar=pd.Series(sayılar,etiketler)
print(pandas_etiketli_sayılar["B"])
```

```python
etiketler=["AB","EU","GB","JP","USA"]
puan1=[10,20,60,40,30]
puan2=[60,30,50,80,62]
veri1=pd.Series(data=puan1,index=etiketler)
veri2=pd.Series(data=puan2,index=etiketler)
print(veri1,veri2)

print("toplam=",veri1+veri2)
```

#### pandas veri yapıları - Veri çerçeveleri -

```python
sayılar=np.arange(0,20).reshape(4,5)
print(sayılar)
print("-"*30) # araya çizgi ekleme
sütunlar=["A","B","C","D","E"]
satırlar=["AB","EU","GB","JP"]
tablo=pd.DataFrame(sayılar,index=satırlar,columns=sütunlar)
print(tablo)
```

#### seçim-ekleme-çıkarma işlemleri

```python
# A sütünundaki verileri gösterme
tablo["A"]
```

```python
# A ve B sütünundaki verileri gösterme
tablo[["A","B"]]
```

```python
# AB satırındaki verileri gösterme
tablo.loc["AB"]
```

```python
tablo.iloc[0]
```

```python
# AB,EU satır ve C,D sütünlarının kesişimini seçmek 
tablo.loc[["AB","EU"],["C","D"]]
```

#### koşullu seçim

 tabloda 2 den büyük sayıları seçmek
 tabloda 2 den büyük ve 10 dan küçük sayıları seçmek

```python
tablo[tablo>2]
```

```python
tablo[(tablo>2) & (tablo<10)]
```

# bazı matematiksel işlemler

```python
print(" Tablodaki en büyük değerler   \n",tablo.max())
print(" Tablodaki ortalama değerler   \n",tablo.mean())
print("C sutünundaki en büyük değer     ",tablo["C"].max())
print("C sutünundaki ortalama değer     ",tablo["C"].mean())
print("C sutünundaki en küçük değer     ",tablo["C"].min())
print("C sutünundaki değerlerin toplamı ",tablo["C"].sum())
```

```python
tablo.loc[["AB","EU"],["C","D"]].sum()
```

```python
# veri çerçevesine yeni sütun ekleme
print("öncesi \n",tablo)
print("-"*30)
tablo["toplam"]=tablo["A"]+tablo["B"]+tablo["C"]+tablo["D"]+tablo["E"]
print("sonrası \n",tablo)
```

```python
# veri çerçevesinden sütun çıkarma
print("öncesi \n",tablo)
print("-"*30)
tablo.drop("toplam",axis=1,inplace=True)
print("sonrası \n",tablo)
```

```python
# veri çerçevesinden satır çıkarma
print("öncesi \n",tablo)
print("-"*30)
tablo.drop("JP",axis=0,inplace=True)
print("sonrası \n",tablo)
```

#### dağılım grafikleri

```python
veriler=sns.load_dataset("exercise")
veriler.head()
```

```python
sns.displot(veriler)
```

```python
sns.jointplot(x='time',y='pulse',data=veriler,kind='scatter')  
# “scatter”  “reg” “resid” “kde” “hex”
```

```python
sns.pairplot(veriler)
```

```python
sns.pairplot(veriler, hue='diet',palette='coolwarm')
```

```python
sns.kdeplot(veriler["pulse"])
```

#### matrix plot


```python
# yıl ve aylara göre uçuş sayıları
uçuşlar = sns.load_dataset('flights')
uçuşlar.corr()
```

```python
#korelasyon verilerine ait ısı haritası
sns.heatmap(uçuşlar.corr())
```

```python
# uçuş verilerine ait ısı haritası oluşturmak için pivot tablosu
pivot_tablosu=uçuşlar.pivot_table(values='passengers',index='month',columns='year')
print(pivot_tablosu)
```

```python
sns.heatmap(pivot_tablosu)
```

```python
# bazı görsel özellikler
sns.heatmap(pivot_tablosu,cmap='magma',linecolor='white',linewidths=1)
```

```python
# kümeleme haritası 
sns.clustermap(pivot_tablosu)
```

#### ızgara grafikler


```python
# iris çiçeği özellikleri veri seti oldukça yaygın bilinn bir veri seti
iris_verileri = sns.load_dataset('iris')
iris_verileri.head()
```

```python
# pairplot
sns.pairplot(iris_verileri)
```

```python
sns.pairplot(iris_verileri,hue='species',palette='rainbow')  # gruplara ayırma ve renk paleti seçimi
```

```python
sns.PairGrid(iris_verileri)# ızgara şablonu oluştu grafikler için çizim türü eklemek gerekiyor
```

```python
import matplotlib.pyplot as plt

ızgara=sns.PairGrid(iris_verileri)
ızgara.map(plt.scatter)
```

```python
# tahmin (regresyon)grafikleri
uçuşlar = sns.load_dataset('flights')
uçuşlar.head()
```

```python
sns.lmplot(x="year",y="passengers",data=uçuşlar)
```

```python
sns.lmplot(x="year",y="passengers",data=uçuşlar,hue="month")
```

```python
sns.lmplot(x="year",y="passengers",data=uçuşlar,col="month") # her ay için farklı bir grafik
```
